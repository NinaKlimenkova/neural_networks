```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Генерація набору даних

```python
X, Y = make_regression (
    n_samples=100000, # Кількість зразків
    n_features=22, # Кількість ознак
    n_informative=100000, # Кількість використаних зразків
    n_targets=1, # кількість виходів регресії. Скаляр в данному випадку
    bias=0.0001, # Член зміщення в базовій моделі
    effective_rank = None, # Вихід центрований, обумовлений та з гаусовською одиничною дисперсією
    tail_strength=0.0, # Важливість жирного шуму хвоста
    noise=0.3, # Шум
    shuffle=False, # Перемішати
    coef=False, # Коефіцієнти базової лінійної моделі
    random_state=500 # Генерація випадкових чисел для набору.
)

print("X_test = ",X[:1])
print("Y_test = ",Y[:1])
```

# Експорт у csv

```python
#створимо датафрейм, щоб записати його у csv
X_df = pd.DataFrame(X)
Y_df = pd.DataFrame(Y)
dataframe = pd.concat([X_df,Y_df], axis = 1)
dataframe
```

```python
dataframe.to_csv('data.csv')
```
