```python
import numpy as np
from random import randint
import matplotlib.pyplot as plt
import random
from random import random 
```

```python
b = randint(0, 10)
k = randint(0, 10)
x = np.array([np.linspace(0, 10, 50)], ndmin=2).reshape((50,1))
y = k*x+b

plt.title("Лінійна функція y = {}*x+{}".format(k,b))
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(x, y)
```

```python

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
Y_pred = model.predict(y)
```

```python
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error

# calculate errors
errors = list()
for i in range(len(y)):
	# calculate error
	err = (y[i] - Y_pred[i])**2
	# store error
	errors.append(err)
	# report error
	print('>%.1f, %.1f = %.3f' % (y[i], Y_pred[i], err))
# plot errors
pyplot.plot(errors)
pyplot.xticks(ticks=[i for i in range(len(errors))], labels=Y_pred)
pyplot.xlabel('Predicted Value')
pyplot.ylabel('Mean Squared Error')
pyplot.show()
```
